# Hitung Prosentase Nilai Unique
Sebelum kita membuat chart dari kolom-kolom attrition_flag, customer_age,gender, education_level, dan income_category, kita akan menghitung masing-masing prosentase nilai unique dari masing-masing kolom sebagai informasi pendukung

```python
persentase_attrition_flag = df['Attrition_Flag'].value_counts(normalize=True) * 100
persentase_attrition_flag
```

```python
persentase_education_level = df['Education_Level'].value_counts(normalize=True) * 100
persentase_education_level
```
