## Analisis Kolom Storage

```python
## Pertama kita akan melihat nilai unik dari dataframe khususnya pada kolom Storage
df['Storage '].unique()
```

Hasil yang diberikan dari perintah df['Storage '].unique() di atas adalah berupa array dari nilai unik yang ada di kolom Storage. Jika kita melihat bahwa **selain semua tipe data pada kolom Storage masih bertipe object, ada ketidaksamaan isi dari kolom Storage. Dimana ada data yang mengandung kata GB dan tidak. Untuk hal ini maka perlu kita melakukan penghapusan kata GB**. Kode berikut ini adalah contohnya :

```python
s = '128 GB' # Kita definisikan sebuah variabel s dengan nilai string 128<spasi>GB
s = s.replace(' GB', '') # mereplace string '<spasi>GB' dari string variabel s untuk kemudian disimpan kembali dalam variabel s
s # tampilkan isi dari variabel s, hasil yang diberikan adalah 128 saja tanpa GB
```

```python
type(s) # cek tipe data dari variabel s
```

Kode berikut berfungsi untuk menghapus kata GB pada data-data dari kolom Storage yang masih mengandung GB. Selain itu juga kita akan melakukan konversi ke dalam tipe data integer untuk data-data di kolom Storage. Konversi ini dibutuhkan agar ketika dilakukan describe pada kolom Storage, dihasilkan gambaran umum berupa statistik deskriptif, seperti : jumlah (count), rata-rata (mean), deviasi standar (std), nilai minimum (min), persentil ke-25 (Q1), median (persentil ke-50 atau Q2), persentil ke-75 (Q3) dan nilai maksimum (max).

```python
df['Storage '] = df['Storage '].str.replace(' ', '')
df['Storage '] = df['Storage '].str.replace('GB', '')
df['Storage '] = df['Storage '].astype(int)
```

## Analisis Kolom Price

```python
## Pertama kita akan melihat nilai unik dari dataframe khususnya pada kolom Price
df['Price ($)'].unique()
```

Hasil yang diberikan dari perintah df['Price (tanda dollar)'].unique di atas adalah berupa array dari nilai unik yang ada di kolom Price (tanda dollar). Jika kita melihat bahwa selain semua tipe data pada kolom Price (tanda dollar) masih bertipe object, ada ketidaksamaan isi dari kolom Price (tanda dollar). Dimana ada data yang mengandung tanda dollar, koma (,) dan spasi. Untuk hal ini maka perlu kita melakukan replace simbol-simbol tersebut pada semua data-data dari kolom Price (tanda dollar) yang masih mengandung simbol-simbol tersebut. Selain itu juga kita akan melakukan konversi ke dalam tipe data integer untuk data-data di kolom Price (tanda dollar).

```python
df['Price ($)'] = df['Price ($)'].str.replace('$', '')
df['Price ($)'] = df['Price ($)'].str.replace(',', '')
df['Price ($)'] = df['Price ($)'].str.replace(' ', '')
df['Price ($)'] = df['Price ($)'].astype(int)
```

Sekarang kita akan mengecek tipe data Price dengan menggunakan perintah df.info

```python
df.info()
```

Sekarang kita akan mengecek gambaran umum dari dataframe menggunakan df.describe. Disini kita melihat perbedaan dengan hasil describe sebelumnya hanya menampilkan data statistik deskriptif dari kolom Battery Capacity (mAh), Storage dan RAM saja, sedangkan describe hasil yang baru juga menampilkan data statistik deskriptif dari kolom Price

```python
df.describe()
```

# Grouping

```python
# hitung rata2 harga untuk setiap Brand
df.groupby('Brand').agg(rata2_harga = ('Price ($)', 'mean')).sort_values(by = 'rata2_harga')
```

```python
df.groupby('Brand').agg(rata2_harga = ('Price ($)', 'mean'), rata2_battery_dollar = ('Battery dollar', 'mean')).\
    sort_values(by = 'rata2_battery_dollar', ascending = False)
```

# Pivot Table

```python
df.pivot_table(index = 'Brand', values = 'Price ($)', aggfunc = 'mean')
```

# Lambda Function

```python
df
```

```python
def get_category(price):
    category = ''
    if price <= 500:
        category = 'MURAH'
    else:
        category = 'MAHAL'
    return category
```

```python
get_category(1000)
```

```python
for i in range(len(df)):
    df.loc[i, 'category'] = get_category(df.loc[i, 'Price ($)'])
```

```python
df
```

```python
df['category_lambda'] = df['Price ($)'].apply(lambda x: 'MURAH' if x <= 500 else ('SEDANG' if x<= 1000 else 'MAHAL'))
```
