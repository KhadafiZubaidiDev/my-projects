# Jawaban No. 1
Soal : [50 poin] Buatlah 5 visualisasi masing-masing untuk kolom attrition_flag, customer_age,gender, education_level, dan income_category. (Wajib ada pie chart, donut chart, vertical bar chart, dan horizontal bar chart).

### Chart Kolom Attrition Flag

#### Pie Chart

```python
fig,ax = plt.subplots()
ax.pie(df['Attrition_Flag'].value_counts(), explode={0,0.1},labels=df['Attrition_Flag'].value_counts().index,autopct='%1.1f%%',colors=['gold','mistyrose'],shadow=True,textprops={'size': 'smaller'})
ax.axis('equal')
ax.legend()
ax.set_title('Distribusi Attrition Flag')
```

#### Donut Chart

```python
fig, ax = plt.subplots(figsize=(10, 6), subplot_kw=dict(aspect="equal"))
wedges, texts = ax.pie(df['Attrition_Flag'].value_counts(), wedgeprops=dict(width=0.7, edgecolor='w'), colors=['gold','mistyrose'])
# Inner circle
centre_circle = plt.Circle((0,0),0.40,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Add labels
labels = df['Attrition_Flag'].value_counts().index + " ( " + persentase_attrition_flag.astype(str) + " %)"
ax.legend(wedges, labels,
      title="Attrition Flag",
      loc="center left",
      bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(texts, size=12, weight="bold")

ax.set_title("Distribusi Attrition Flag")
ax.axis('equal')

plt.show()
```

#### Vertical Bar Chart

```python
ax = df['Attrition_Flag'].value_counts().plot(kind='bar',color=['gold','mistyrose'], figsize=(15,15))
ax.set_title('Distribusi Attrition Flag')
ax.set_ylabel('Jumlah')
ax.set_xlabel('Attrition Flag')
plt.show()
```

#### Horizontal Bar Chart

```python
ax = df['Attrition_Flag'].value_counts().plot(kind='barh',color=['gold','mistyrose'], figsize=(15,15))
ax.set_title('Distribusi Attrition Flag')
ax.set_xlabel('Jumlah')
ax.set_ylabel('Attrition Flag')
plt.show()
```

### Chart Kolom Customer Age

#### Pie Chart

```python
fig,ax = plt.subplots()
ax.pie(df['Customer_Age'].value_counts(), labels=df['Customer_Age'].value_counts().index)
ax.axis('equal')
ax.legend()
ax.set_title('Customer Age')
```

#### Donut Chart

```python
fig, ax = plt.subplots(figsize=(10, 6), subplot_kw=dict(aspect="equal"))
wedges, texts = ax.pie(df['Customer_Age'].value_counts(), wedgeprops=dict(width=0.7, edgecolor='w'))
# Inner circle
centre_circle = plt.Circle((0,0),0.40,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Add labels
labels = df['Customer_Age'].value_counts().index
ax.legend(wedges, labels,
      title="Customer Age",
      loc="center left",
      bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(texts, size=12, weight="bold")

ax.set_title("Distribusi Customer Age")
ax.axis('equal')

plt.show()
```

#### Vertical Bar Chart

```python
ax = df['Customer_Age'].value_counts().plot(kind='bar', figsize=(15,15))
ax.set_title('Distribusi Customer Age')
ax.set_ylabel('Jumlah')
ax.set_xlabel('Customer Age')
plt.show()
```

#### Horizontal Bar Chart

```python
ax = df['Customer_Age'].value_counts().plot(kind='barh',figsize=(15,15))
ax.set_title('Distribusi Customer Age')
ax.set_xlabel('Jumlah')
ax.set_ylabel('Customer Age')
plt.show()
```
