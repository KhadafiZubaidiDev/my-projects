## Deskripsi Project
Project ini merupakan salah satu tugas Mini Project pada Kelas Data Science & AI Fundamentals Batch 42 yang diselenggarakan oleh DigitalSkola. Mini Project ini dikerjakan oleh : Khadafi Zubaidi.

Dalam project ini dijelaskan mengenai langkah-langkah untuk membersihkan data, menampilkan data, serta melakukan analisis data eksploratif menggunakan bahasa pemrograman Python dan file data berekstensi csv.
