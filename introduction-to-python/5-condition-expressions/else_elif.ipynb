```python
x = 28

if x < 0:
    print('x < 0')                     # Executes only if x < 0
elif x == 0:
    print('x is zero')                 # If it's not true that x < 0, check if x == 0
elif x == 1:
    print('x == 1')                    # If it's not true that x < 0 and x != 0, check if x == 1
else:
    print('none of the above is true')

name = "John"

if name == "John":
    print(True)
else:
    print(False)
```
